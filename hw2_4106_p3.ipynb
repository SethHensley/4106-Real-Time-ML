```python
#seth hensley
#801055921
#problem 3

import tensorflow as tf
import keras 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
#attempting to write lanet from class with tf inclding the changes the hw calls for 


def build_lanet1():
    model = tf.keras.Sequential([
        tf.keras.Input(shape = (28,28,1)),
        keras.layers.Conv2D(6, 2, padding = "same"),
        keras.layers.MaxPool2D(2, strides = 2),
        keras.layers.Dropout(0.15),
        keras.layers.Conv2D(12, 2),
        keras.layers.Activation("relu"),
        keras.layers.MaxPool2D(2, strides = 2),
        keras.layers.Dropout(0.15),        
        keras.layers.Flatten(),
        keras.layers.Dense(120),
        keras.layers.Activation("relu"),
        keras.layers.Dropout(0.2),
        keras.layers.Dense(84),
        keras.layers.Activation("relu"),
        keras.layers.Dropout(0.2),
        keras.layers.Dense(10)       
        
        
    ])
    
    
    return model


def build_lanet2():
    model = tf.keras.Sequential([
        tf.keras.Input(shape = (28,28,1)),
        keras.layers.Conv2D(12, 6, padding = "same"),
        keras.layers.MaxPool2D(2, strides = 2),
        keras.layers.Dropout(0.15),
        keras.layers.Conv2D(24, 6),
        keras.layers.Activation("relu"),
        keras.layers.MaxPool2D(2, strides = 2),
        keras.layers.Dropout(0.15),
        keras.layers.Flatten(),
        keras.layers.Dense(240),
        keras.layers.Activation("relu"),
        keras.layers.Dropout(0.2),
        keras.layers.Dense(168),
        keras.layers.Activation("relu"),
        keras.layers.Dropout(0.2),
        keras.layers.Dense(10)       
        
        
    ])
    
    
    return model



def build_lanet3():
    model = tf.keras.Sequential([
        tf.keras.Input(shape = (28,28,1)),
        keras.layers.Conv2D(6, 6, padding = "same"),
        keras.layers.MaxPool2D(2, strides = 2),
        keras.layers.Dropout(0.15),
        keras.layers.Conv2D(12, 6),
        keras.layers.Activation("relu"),
        keras.layers.Dropout(0.15),
        keras.layers.Conv2D(24,6,padding="same"),
        keras.layers.Activation("relu"),
        keras.layers.Dropout(0.15),
        keras.layers.Conv2D(12,6,padding="same"),
        keras.layers.Activation("relu"),
        keras.layers.MaxPool2D(2, strides = 2),
        keras.layers.Dropout(0.15),
        keras.layers.Flatten(),
        keras.layers.Dense(120),
        keras.layers.Activation("relu"),
        keras.layers.Dropout(0.2),
        keras.layers.Dense(84),
        keras.layers.Activation("relu"),
        keras.layers.Dropout(0.2),
        keras.layers.Dense(10)       
        
        
    ])
    
    
    return model



def build_lanet4():
    model = tf.keras.Sequential([
        tf.keras.Input(shape = (28,28,1)),
        keras.layers.Conv2D(6, 6, padding = "same"),
        keras.layers.MaxPool2D(2, strides = 2),
        keras.layers.Dropout(0.15),
        keras.layers.Conv2D(12, 6),
        keras.layers.Activation("relu"),
        keras.layers.MaxPool2D(2, strides = 2),
        keras.layers.Dropout(0.15),
        keras.layers.Flatten(),
        keras.layers.Dense(120),
        keras.layers.Activation("relu"),
        keras.layers.Dropout(0.2),
        keras.layers.Dense(128),
        keras.layers.Activation("relu"),
        keras.layers.Dropout(0.2),
        keras.layers.Dense(84),
        keras.layers.Activation("relu"),
        keras.layers.Dropout(0.25),
        keras.layers.Dense(42),
        keras.layers.Activation("relu"),
        keras.layers.Dropout(0.2),
        keras.layers.Dense(10)       
        
        
    ])
    
    
    return model

```

```python
(xtrain,ytrain),(xtest,ytest) = keras.datasets.fashion_mnist.load_data()
print("xtrain shape:", xtrain.shape, "ytrain shape:", ytrain.shape)
```

```python
model1 = build_lanet1()
model2 = build_lanet2()
model3 = build_lanet3()
model4 = build_lanet4()

```

```python
model1.compile(optimizer="adam", loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits= True), metrics=["accuracy"])
model1.summary()

model2.compile(optimizer="adam", loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits= True), metrics=["accuracy"])
model2.summary()

model3.compile(optimizer="adam", loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits= True), metrics=["accuracy"])
model3.summary()

model4.compile(optimizer="adam", loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits= True), metrics=["accuracy"])
model4.summary()
```

```python
model1.fit(xtrain,ytrain,validation_data=(xtest,ytest),epochs=25)
model2.fit(xtrain,ytrain,validation_data=(xtest,ytest),epochs=25)
model3.fit(xtrain,ytrain,validation_data=(xtest,ytest),epochs=25)
model4.fit(xtrain,ytrain,validation_data=(xtest,ytest),epochs=25)
```
